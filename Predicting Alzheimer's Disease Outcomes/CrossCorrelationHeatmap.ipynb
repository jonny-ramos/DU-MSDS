```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('oasis_longitudinal.csv')
df.head()
```

```python
sns.pairplot(df, hue='Group')
plt.savefig('dementia_long_pairplot.png', dpi=600)
```

```python
sns.heatmap(df.corr(), cmap='bwr', vmin=-1, vmax=1)
plt.axis('equal')
plt.savefig('dementia_long_heatmap_corr_alldata.png', dpi=600)
```

```python
sns.heatmap(df[df['Group'] == 'Demented'].corr(), cmap='bwr', vmin=-1, vmax=1)
plt.axis('equal')
plt.savefig('dementia_long_heatmap_corr_demented.png', dpi=600)
```

```python
sns.heatmap(df[df['Group'] == 'Nondemented'].corr(), cmap='bwr', vmin=-1, vmax=1)
plt.axis('equal')
plt.savefig('dementia_long_heatmap_corr_nondemented.png', dpi=600)
```

```python
sns.heatmap(df[df['Group'] == 'Converted'].corr(), cmap='bwr', vmin=-1, vmax=1)
plt.axis('equal')
plt.savefig('dementia_long_heatmap_corr_converted.png', dpi=600)
```
